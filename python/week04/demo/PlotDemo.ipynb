```python
import pandas as pd
from sklearn import datasets
import numpy as np

df = datasets.load_iris()

df1 = pd.DataFrame(data= np.c_[df['data'], df['target']],
                     columns= df['feature_names'] + ['FlowerName'])
pd.plotting.parallel_coordinates(df1, "FlowerName", color=["red", "yellow","blue"])
```

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
  
# selecting three numerical features


#add csv file to dataframe
df = pd.read_csv('hp.txt')
features = ['Number of Bedrooms', 'Size(Sq M)',
            'Price']
   
# plotting the scatter matrix
# with the features
scatter_matrix(df[features])
plt.show()
```
